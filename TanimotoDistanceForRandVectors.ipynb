```python
import numpy as np

tanimotos = []

size = 166

for x in range(10000):
    vector1 = np.random.randint(0,2,size=(size))
    vector2 = np.random.randint(0,2,size=(size)) # ith your values for the second vector

#     print(vector1)

    dot_product = np.dot(vector1, vector2)
    sum_squares_v1 = np.sum(np.square(vector1))
    sum_squares_v2 = np.sum(np.square(vector2))

    tanimotos.append( dot_product / (sum_squares_v1 + sum_squares_v2 - dot_product))
```

```python
plt.figure()
plt.hist(tanimotos)
```
